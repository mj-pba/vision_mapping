```python
import os 
import numpy as np
import matplotlib.pyplot as plt
```

```python
current_working_directory = os.getcwd()

# Move one directory back
parent_directory = os.path.dirname(current_working_directory)
print(parent_directory)

#data file parth 
file_path = os.path.join(parent_directory, 'data')
print(file_path)

# data file path and name
file_path_name = os.path.join(file_path, '20240620_5cycle_22_stops_400um_step_combined.csv')
print(file_path_name)

```

```python
# Load the CSV file into a numpy array, skipping the first N rows
N = 1  # Number of rows to skip
data = np.genfromtxt(file_path_name, delimiter=',', skip_header=N)

# Load the headers separately
with open(file_path_name, 'r') as file:
    headers = file.readline().strip().split(',')

# Set print options to suppress scientific notation
np.set_printoptions(suppress=True,precision=6)

print(data.shape)
```

```python
for x in range(0,data.shape[1],1):
    print(x ," - ",headers[x]," - ", data[0][x])
```

```python
print(headers[2])

for x in range(0,data.shape[0],1):
    print(x ," - ", data[x][2])
```

```python
print(data[:,2])
```

```python
# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(data[:,2], data[:,3], marker='1', color='b',label = headers[3])
plt.plot(data[:,2], data[:,4]/2.2 + 2, marker='2', color='g',label=headers[4])

plt.plot(data[:,2], data[:,6], marker='1', color='b',label = headers[6])
plt.plot(data[:,2], data[:,7]/2.2, marker='2', color='g',label=headers[7])

# Add labels and title
plt.xlabel('Axis Position (mm)')
plt.ylabel('Position Error (μm)')
plt.title('Position Error vs. Axis Position: Comparing Camera Capability Based on Laser Values')

plt.legend()
# Display the plot
plt.grid(True)
plt.show()
```
