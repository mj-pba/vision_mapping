```python
import matplotlib.pyplot as plt
import cv2
import os
import numpy as np

# Iterate through each file and plot individually
for file_name in png_files:
    image_path = os.path.join(folder_name, file_name)
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    if image is None:
        print(f"Could not read {file_name}")
        continue

    # --- Calculations ---
    # 1. Parse Filename for Illumination and Exposure
    # Assuming format: Illumination_Exposure.png (e.g., 50_1400.png)
    try:
        name_parts = os.path.splitext(file_name)[0].split('_')
        illumination_intensity = name_parts[0]
        exposure_time = name_parts[1] if len(name_parts) > 1 else "N/A"
    except:
        illumination_intensity = "Unknown"
        exposure_time = "Unknown"

    # 2. Intensity Max and Min
    intensity_max = np.max(image)
    intensity_min = np.min(image)

    # 3. Feature Contrast Ratio (CR)
    # CR = (I_feature - I_background) / (I_feature + I_background)
    # Using Max as Feature and Min as Background
    if (int(intensity_max) + int(intensity_min)) != 0:
        contrast_ratio = (int(intensity_max) - int(intensity_min)) / (int(intensity_max) + int(intensity_min))
    else:
        contrast_ratio = 0.0

    # 4. Reflectivity Uniformity
    # "Consistency of feature intensity... Maximum intensity variation"
    # We estimate features as pixels above a threshold (e.g., mid-point of dynamic range) to exclude background
    threshold = (int(intensity_max) + int(intensity_min)) / 2
    feature_pixels = image[image > threshold]
    
    if len(feature_pixels) > 0:
        feat_max = np.max(feature_pixels)
        feat_min = np.min(feature_pixels)
        # Calculate variation percentage: (Max - Min) / Max * 100
        reflectivity_uniformity = ((feat_max - feat_min) / feat_max) * 100
    else:
        reflectivity_uniformity = 0.0

    # Print Results
    print(f"--- Image: {file_name} ---")
    print(f"Illumination Intensity: {illumination_intensity}")
    print(f"Exposure Time (us):     {exposure_time}")
    print(f"Intensity Max:          {intensity_max}")
    print(f"Intensity Min:          {intensity_min}")
    print(f"Contrast Ratio (CR):    {contrast_ratio:.4f} (Target > 0.6 approx 4:1)")
    print(f"Treserhold for Features:  {threshold:.2f}")
    print(f"Reflectivity Uniformity:{reflectivity_uniformity:.2f}% (Target < 5%)")
    print("-" * 30)

    # --- Plotting ---
    # Extract data
    row_pixels = image[1500, :]
    col_pixels = image[:, 2048]

    # Create a new figure for this specific image
    plt.figure(figsize=(15, 5))
    
     # Subplot 1: Row Variation
    plt.subplot(1, 2, 1)
    plt.plot(row_pixels)
    plt.title(f'File: {file_name} | Row 1500')
    plt.xlabel('Column Index')
    plt.ylabel('Pixel Value')
    plt.ylim(0, 260)  # Fixed range for comparison
    plt.grid(True, alpha=0.3)
    

    # Subplot 2: Column Variation
    plt.subplot(1, 2, 2)
    plt.plot(col_pixels, color='orange')
    plt.title(f'File: {file_name} | Column 2048')
    plt.xlabel('Row Index')
    plt.ylabel('Pixel Value')
    plt.ylim(0, 260)  # Fixed range for comparison
    plt.grid(True, alpha=0.3)
```

```python
import matplotlib.pyplot as plt
import cv2
import os
import numpy as np

# Iterate through each file and plot individually
for file_name in png_files:
    image_path = os.path.join(folder_name, file_name)
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    if image is None:
        print(f"Could not read {file_name}")
        continue

    # --- Calculations ---
    # 1. Parse Filename for Illumination and Exposure
    # Assuming format: Illumination_Exposure.png (e.g., 50_1400.png)
    try:
        name_parts = os.path.splitext(file_name)[0].split('_')
        illumination_intensity = name_parts[0]
        exposure_time = name_parts[1] if len(name_parts) > 1 else "N/A"
    except:
        illumination_intensity = "Unknown"
        exposure_time = "Unknown"

    # 2. Intensity Max and Min

    row_pixels = image[1500, :]
    col_pixels = image[:, 2048]

    intensity_max = np.max(row_pixels)
    intensity_min = np.min(row_pixels)

    threshold = (int(intensity_max) + int(intensity_min)) / 2
    # get max intensity values as array
    max_intensity_pixels = row_pixels[row_pixels > threshold]
    min_intensity_pixels = row_pixels[row_pixels <= threshold]
    average_max_intensity = np.mean(max_intensity_pixels) if len(max_intensity_pixels) > 0 else 0
    average_min_intensity = np.mean(min_intensity_pixels) if len(min_intensity_pixels) > 0 else 0

    standard_deviation_max = np.std(max_intensity_pixels) if len(max_intensity_pixels) > 0 else 0
    standard_deviation_min = np.std(min_intensity_pixels) if len(min_intensity_pixels) > 0 else 0
    


    # 3. Feature Contrast Ratio (CR)
    # CR = (I_feature - I_background) / (I_feature + I_background)
    # Using Max as Feature and Min as Background
    if (average_max_intensity + average_min_intensity) != 0:
        contrast_ratio = (average_max_intensity - average_min_intensity) / (average_max_intensity + average_min_intensity)
        #contrast_ratio = average_max_intensity/average_min_intensity
    else:
        contrast_ratio = 0.0

    # 4. Reflectivity Uniformity
    # "Consistency of feature intensity... Maximum intensity variation"
    # We estimate features as pixels above a threshold (e.g., mid-point of dynamic range) to exclude background

    feature_pixels = image[image > threshold]
    
    if len(feature_pixels) > 0:
        feat_max = np.max(feature_pixels)
        feat_min = np.min(feature_pixels)
        # Calculate variation percentage: (Max - Min) / Max * 100
        reflectivity_uniformity = ((feat_max - feat_min) / feat_max) * 100
    else:
        reflectivity_uniformity = 0.0

    # Print Results
    print(f"--- Image: {file_name} ---")
    print(f"Illumination Intensity: {illumination_intensity}")
    print(f"Exposure Time (us):     {exposure_time}")
    print(f"Intensity Max:          {average_max_intensity:.2f}")
    print(f"Intensity Min:          {average_min_intensity:.2f}")
    print(f"Intensity dif:          {average_max_intensity - average_min_intensity:.2f}")
    
    print(f"Standard Deviation Max:  {standard_deviation_max:.2f}")
    print(f"Standard Deviation Min:  {standard_deviation_min:.2f}")
    print(f"Contrast Ratio (CR):    {contrast_ratio:.4f} (Target > 0.6 approx 4:1)")
    print(f"Treserhold for Features:  {threshold:.2f}")
    print(f"Reflectivity Uniformity:{reflectivity_uniformity:.2f}% (Target < 5%)")
    print("-" * 30)
```
