### 1. Theoretical Analysis (What We've Been Discussing)

This is the analysis done before running the code, focusing on the algorithm's design.
**What it measures:** The number of operations the algorithm will perform relative to the input size ('n'). This is expressed using Big O notation (e.g., O(n), O(log n)).

**Why we use it:** It provides a universal and hardware-independent way to judge an algorithm's efficiency. An algorithm with O(n) complexity will be fundamentally faster for large inputs than an O(n²) algorithm, regardless of whether you run it on a supercomputer or a ten-year-old laptop. It allows us to compare the scalability of different algorithms.

Goal: To understand how the algorithm will behave as the input size grows to infinity. This helps in choosing the right algorithm for a problem before a single line of code is written.

### 2. Practical Performance Measurement (Profiling)

This is the analysis done while the code is executing.
**What it measures:** The actual, real-world wall-clock time and the amount of memory (in megabytes, gigabytes, etc.) consumed by the program.

**Why we use it:** To find bottlenecks: Profiling tools can show you exactly which parts of your code are taking the most time or using the most memory. This is crucial for optimization.
To verify theoretical analysis: It helps confirm that your implementation of the algorithm performs as expected.
For specific environments: It's essential when you need to know how your code will run on a specific piece of hardware or under specific conditions.
Goal: To optimize a specific implementation of an algorithm in a specific environment.

```python
# read file performance_profile_log.csv

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
# data collected from removeing pixel summation and 
test_0_file = 'test0_performance_profile_log.csv'
test_1_file = 'test1_performance_profile_log.csv'
test_2_file = 'test2_performance_profile_log.csv'
test_3_file = 'test3_performance_profile_log.csv'

test_0_data = pd.read_csv(test_0_file)
test_1_data = pd.read_csv(test_1_file)
test_2_data = pd.read_csv(test_2_file)
test_3_data = pd.read_csv(test_3_file)
```

```python
# print(data.head())

# Plot Execution Time vs Images Used for Test
plt.figure(figsize=(10, 6))
plt.plot(test_0_data['Images Used for Test'], test_0_data['Execution Time (ms)']/1000, marker='o', label='Original algorithm')
plt.plot(test_1_data['Images Used for Test'], test_1_data['Execution Time (ms)']/1000, marker='o', label='Optimized algorithm without pixel summation')
plt.plot(test_2_data['Images Used for Test'], test_2_data['Execution Time (ms)']/1000, marker='o', label='Optimized algorithm without print statements')
plt.plot(test_3_data['Images Used for Test'], test_3_data['Execution Time (ms)']/1000, marker='o', label='New halcon calculation optimization')
plt.title('Execution Time vs Images Used for Test')
plt.xlabel('Images Used for Test')
plt.ylabel('Execution Time (s)')
plt.grid()
plt.legend()
plt.show()
```
